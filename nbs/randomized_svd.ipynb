# Randomized SVD

Randomized singular value decomposition (SVD) implementation in Julia.

```julia
using RandNLA
using LinearAlgebra, Random, Plots
```

Simple example with a random matrix.

```julia
A = randn(100, 30);

rank(A)
```

```julia
U, S, Vt = svd(A);
```

```julia
plt = plot(S)
```
